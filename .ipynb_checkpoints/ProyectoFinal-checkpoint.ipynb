```python
# Importar la libreria de pandas :3
import pandas as pd
```

```python
# Cargar el dataset que contiene la información de USA
usa_data = pd.read_csv('/Users/cbqa-0105/Documents/ISSC/Goku/BI_Final/data/usa_county_wise.csv')
```

```python
# Eliminamos las columnas con datos inecesarios
del usa_data['UID']
del usa_data['iso2']
del usa_data['iso3']
del usa_data['code3']
del usa_data['FIPS']
del usa_data['Combined_Key']
```

```python
# Reemplazamos los datos NaN por un espacio en blanco
usa_data["Admin2"].fillna("", inplace = True)
usa_data = usa_data.rename(columns={'Country_Region': 'Country/Region', 'Long_': 'Long', 'Admin2': 'Zone', 'Province_State': 'Province/State'})
```

```python
# Imprimimos usa_data
usa_data
```

```python
covid_clean = pd.read_csv('/Users/cbqa-0105/Documents/ISSC/Goku/BI_Final/data/covid_19_clean_complete.csv')
covid_clean
```

```python
frames = [covid_clean, usa_data]

result = pd.concat(frames)

result["Province/State"].fillna("", inplace = True)
result["Recovered"].fillna(0.0, inplace = True)
result["Zone"].fillna("", inplace = True)

result
```

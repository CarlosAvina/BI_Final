```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Leer archivo
df = pd.read_csv('data/result.csv')
df["Zone"] = df["Zone"].fillna('')
df["Province/State"] = df["Province/State"].fillna('')

df.head()
```

```python
df['Country/Region'].value_counts()
```

```python
#Transforming the string values into 0 or 1 so that we can use them in the training settings 
#We are also gonna drop the first columm because it's essentialy not needed as we can deduct it's 
#contents by observing the content of the othet ones
sex=pd.get_dummies(df['Confirmed'],drop_first=True)
city_cat=pd.get_dummies(df['Deaths'],drop_first=True)
age=pd.get_dummies(df['Recovered'],drop_first=True)
sex
city_cat
age
```

```python
#we need to concatenate the values that we converted to the original table so that we can use them 
#as parte o the training model
df=pd.concat([df,sex,city_cat,age],axis=1)
#we drop the original colummns
df=df.drop(['Confirmed','Deaths','Recovered'],axis=1)
df
```

```python
#We choose the variable that we want to predict
y = df['Country/Region']
X = df.drop(columns = ['Country/Region'])
```

```python
#here we divide the data set 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75, random_state=1)
```

```python
#import the Logistic regression model and apply it 
from sklearn.linear_model import LogisticRegression
logi = LogisticRegression()
#fitting the model
logi.fit(X_train,y_train)
```

```python
#We can finally make the prediction
y_pred = logi.predict(X_test)
```

```python
X_test
cnf_matrix = confusion_matrix(y_test, y_pred)
```

```python
#extracting true_positives, false_positives, true_negatives, false_negatives
#tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
tn = confusion_matrix(y_test, y_pred).ravel()
fp = confusion_matrix(y_test, y_pred).ravel()
fn = confusion_matrix(y_test, y_pred).ravel()
tp = confusion_matrix(y_test, y_pred).ravel()

print("True Negatives: ",tn)
print("False Positives: ",fp)
print("False Negatives: ",fn)
print("True Positives: ",tp)
```

```python
#Accuracy (%) 
Accuracy = (tn+tp)/(tp+tn+fp+fn) 
#print("Accuracy {:0.2f}".format(Accuracy))
Accuracy
```

```python
#Precision 
Precision = tp/(tp+fp) 
#print("Precision {:0.2f}".format(Precision))
Precision
```

```python
#Specificity 
Specificity = tn/(tn+fp)
print("Specificity {:0.2f}".format(Specificity))
```

```python
#ROC (Receiver Operating Characteristic curve)
import scikitplot as skplt #to make things easy

y_pred_proba = logi.predict_proba(X_test)
skplt.metrics.plot_roc_curve(y_test, y_pred_proba)
plt.show()
```
